```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def gradient_discent(x,y):
    m_curr = c_curr = 0
    iterations = 5000
    n = len(x)
    learning_rate = 0.01                              # Chose this value after few experimentations
    for i in range(iterations):
        y_pred =  m_curr*x + c_curr
        mse= 1/n*sum([val**2 for val in (y-y_pred)])  # mean square error
        md = 2/n*sum(-x*(y-y_pred))                   # derivative of mse w.r.t. m 
        cd = 2/n*sum(-(y-y_pred))                     # derivative of mse w.r.t. c
        m_curr = m_curr - learning_rate*md
        c_curr = c_curr - learning_rate*cd
        plt.plot(x,y_pred)
        print("m {}, c {}, mse {}, iterations {},".format(m_curr,c_curr,mse,i))
gradient_discent(x,y)
plt.plot(x,y)
plt.show()
```
